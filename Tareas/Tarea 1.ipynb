Dado un conjunto de datos cuantitativos, podemos obtener su promedio matemático calculando la suma de todos sus elementos y dividiendola por el número total de los mismos. Así, si tenemos un conjunto con n elementos, calculamos su promedio como
$$\dfrac{1+2+...+n}{n}$$

## Ejemplo:
Dado el conjunto A=(1,2,3,4,5), calculamos su promedio como
$$\dfrac{1+2+3+4+5}{5} = \dfrac{15}{5} = 3$$
